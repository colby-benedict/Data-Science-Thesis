## 1.1 Background

According to the Centers for Disease Control and Prevention (CDC), “Lyme disease is the most common vector-borne disease in the United States." [1] Living organisms that can transmit infectious pathogens, such as the pathogens that cause Lyme Disease, between humans, animals, or both are vectors. Some common vectors of disease include ticks, mosquitoes, and fleas. [2] Lyme Disease is known to be spread by certain hard-bodied ticks of genus Ixodes. According to the CDC, there is no credible evidence that Lyme disease can be transmitted through air, food, water, or from direct person-to-person contact, dogs or cats with Lyme disease directly to their owners, or bites by mosquitoes, flies, fleas, or lice. It is possible, but rare for a mother with untreated Lyme disease during pregnancy to have an infected placenta and spread the disease to the fetus. There is no increased risk of adverse birth outcomes when the pregnant mother with Lyme disease has an appropriate antibiotic treatment. There have been no cases of Lyme disease being linked to blood transfusion, but the Lyme disease bacteria has been found able to live in blood that is stored for donation. [3]

Lyme disease, also called Lyme borreliosis, is  a tick-borne disease caused by the infection of bacteria of the genus Borrelia. Lyme disease is primarily caused by the bacterium Borrelia burgdorferi which is found in North America and Europe, Borrelia afzelii which is found in Europe and Asia, and Borrelia garinii which is also found in Europe and Asia. [6] Lyme disease predominantly occurs in temperate regions of the northern hemisphere. [4] In North America, Lyme disease is only known to be caused by Borrelia burgdorferi and Borrelia mayonii. Borrelia mayonii is a newer recognized and much less common Lyme disease causing species than Borrelia burgdorferi. [5,1]

The blacklegged tick Ixodes scapularis, commonly known as the deer tick, is the predominant vector that spreads Lyme disease in the Northeast and Midwest regions of the United States and some areas in middle southern and southeastern Canada. The western blacklegged tick Ixodes pacificus is the predominant vector that spreads Lyme disease on the United States Pacific Coast and West region of the United States. [3,4] Borrelia mayonii is known to be transmitted by the blacklegged tick Ixodes scapularis and found in the upper Midwest region of the United States. [6] Both Ixodes scapularis and Ixodes pacificus, the main tick vectors in North America, transmit Borrelia burgdorferi. [4] The two tick vectors in Europe known to transmit Lyme disease are Ixodes ricinus and Ixodes persulcatus. The main Lyme disease vector in Europe is Ixodes ricinus. While Ixodes ricinus spreads Lyme disease throughout Europe, Ixodes persulcatus spreads Lyme disease in eastern Europe. Ixodes ricinus ticks are known to transmit Borrelia burgdorferi. The tick Ixodes persulcatus is also found in western Russia, the Baltic countries, parts of Finland, central regions of eastern Russia, northern Mongolia, China, and Japan. Ixodes persulcatus transmits Borrelia afzelii and Borrelia garinii, but is not known to transmit Borrelia burgdorferi. [4] "[T]ick abundance and infection prevalence vary widely and are influenced by microclimate, vegetation and the abundance of reservoir vertebrate hosts." [4] The region of the United States with the most intense transmission of B. burgdorferi is the Northeast, "where the prevalence of B. burgdorferi infection among ticks can be as high as 40-50%", whereas, "infection prevalence in the southern United States is less than 1%." The CDC states "Ticks live in grassy, brushy, or wooded areas, or even on animals." [11] Activities done outside such as walking a dog, camping, gardening, and hiking could bring people in close contact with ticks. "Many people get ticks in their own yard or neighborhood." [11] Walking in wooded and brushy areas with high grass and leaf litter may also increase the chance of a tick climbing aboard since ticks wait for a host on the tips of grasses and shrubs, and many ticks, such as blacklegged ticks, wait in a position known as “questing”. [11] Ticks can’t fly or jump, but while questing they hold onto leaves and grass by their lower, or third and fourth, pair of legs and hold the upper, or first, pair of legs outstretched, waiting to climb onto the host. When a host brushes the spot where a tick is waiting, it quickly climbs aboard then finds a suitable place to bite its host. [12, 3]



Once a tick has found its feeding spot on a host's body, it grasps onto the skin, cutting into the surface, then inserts its feeding tube into the host. Not only can the feeding tube have barbs that help keep the tick in place, ticks can also secrete small amounts of saliva with anesthetic properties so that its host cannot feel that the tick has attached itself. 
Ticks can attach to any part of the human body, but prefer warm, moist areas, often found in the groin, armpits, and scalp which can be difficult to see. Consequently, a tiny tick can easily go unnoticed. Unsurprisingly, most humans become infected with the Lyme disease bacteria through the bites of nymphs as they are less than 2 mm in size. Ticks in this stage feed during the spring and summer, while the adult Ixodes ticks are most active during the cooler months of the year any time temperatures are above freezing. Adult ticks can also transmit Lyme disease bacteria, but they are much larger, thus, more likely to be discovered and removed before they have had time to transmit the bacteria. A tick will suck the blood slowly for several days if they are not removed. In most cases, the tick must be attached for at least 36 to 48 hours before the Lyme disease bacterium can be transmitted. After feeding, most ticks will drop off and prepare for the next life stage. [3]

While ticks are sucking on an infected host's (human or animal) blood, they ingest disease-producing microorganisms from their host which they can later transmit into a new host during their subsequent meal as small amounts of saliva from the tick can enter the skin of the host during the feeding process [12]. 
Once a vector becomes infectious, they are often capable of transmitting the pathogen for the rest of their life to whomever's blood they feed on.

After the tick eggs hatch, they must feed on blood at every stage to survive and need a new host at each stage of their life. These hosts include mammals, birds, reptiles, and amphibians. "Reports of Lyme disease are collected and verified by state and local health departments in accordance with their legal mandate and surveillance practices." [22] "Final annual case counts are published when the year is over and all states and territories have verified their data, typically in the fall of the following year." [22] How the data is collected, the rules of collection, and Lyme disease case definitions which have changed over the years can be found on the CDC's website with the following link:
https://wwwn.cdc.gov/nndss/conditions/lyme-disease/
. "Lyme disease has been a nationally notifiable condition in the United States since 1991." Following its implementation in 1991, the national surveillance case definition for Lyme disease was modified in 1996, 2008, 2011, and 2017. The changes were generally minor, but may have had some impact on surveillance. [22] The dataset's 2000-2007 case counts reflect the surveillance case definition in place during that time. "Data from 2008 to the present reflect the case definition which allowed for reporting of "confirmed" and "probable" cases. County codes of "999" represent "unknown" county of residence within each state. [22]


Some limitations of surveillance data (as described by the CDC):
<blockquote>

1. Not every case of Lyme disease is reported to the CDC, and some cases that are reported may be due to another cause. Under-reporting is more likely to occur in highly endemic areas, whereas over-reporting is more likely to occur in non-endemic areas.
2. Surveillance data are subject to each state’s abilities to capture and classify cases, which are dependent upon budget and personnel and varies not only between states, but also from year to year within a given state. Consequently, a sudden or marked change in reported cases does not necessarily represent a true change in disease incidence, and should not be construed as such without knowledge of that state’s historical surveillance practices.
3. Surveillance data are captured by county of residence, not county of exposure.
4. States may close their annual surveillance dataset at a different time than CDC. Thus, the final case counts published by CDC may not exactly match numbers published by each state agency for a given year. [22]

</blockquote>

###  2.1.2 Aggregation of Cases in Three-Year Time Periods

I aggregated cases in three-year time periods (e.g., 2000-2002, 2003-2005, 2006-2009). The aggregated data did not include the "unknowns".

```python
per3_df = pd.DataFrame(LDcounts_cty_df.iloc[:,0:4])

per3_df['Cases2000-2002'] = LDcounts_cty_df.loc[:,LDcounts_cty_df.columns[4:7]].sum(axis=1)
per3_df['Cases2003-2005'] = LDcounts_cty_df.loc[:,LDcounts_cty_df.columns[7:10]].sum(axis=1)
per3_df['Cases2006-2008'] = LDcounts_cty_df.loc[:,LDcounts_cty_df.columns[10:13]].sum(axis=1)
per3_df['Cases2009-2011'] = LDcounts_cty_df.loc[:,LDcounts_cty_df.columns[13:16]].sum(axis=1)
per3_df['Cases2012-2014'] = LDcounts_cty_df.loc[:,LDcounts_cty_df.columns[16:19]].sum(axis=1)
per3_df['Cases2015-2017'] = LDcounts_cty_df.loc[:,LDcounts_cty_df.columns[19:22]].sum(axis=1)

per3_df
```

### 3.3.2 Creation of Lyme Disease Case Incidence and Count Classes

The CDC defines low incidence of Lyme disease cases as the average incidence of less than 10 confirmed cases per 100,000 persons for the previous three reporting years. The CDC defines high incidence of Lyme disease cases as the average incidence of at least 10 confirmed cases per 100,000 persons for previous three reporting years. [31] For purposes of this project, low incidence of Lyme disease cases will be defined as the frequency of less than 10 reported cases per 100,000 persons in one year. High incidence of Lyme disease cases will be defined as the frequency of at least 10 reported cases per 100,000 persons in one year.

I seperated the frequency of Lyme disease into two classes: low incidence and high incidence.

Frequency: 10 cases/100000 persons = 0.0001 cases/person

First I made a copy of the clean Lyme disease dataset that contained all the variables.

```python
LD_full_df = LD_clean_df.copy()
# or could do:
#LD_full_df = pd.read_csv("LD_clean_df.csv", encoding='latin-1')
```

The function displayed in the code below created the classes for the Incidence target variable. The function was then applied to the full Lyme disease dataset where the values in the Incidence column described the Lyme disease incidence for the year in each values' respective row.

```python
def funcInc(row):
    if row['Frequency'] < 0.0001:
        val = 'Low'
    else:
        val = 'High'
    return val
```

```python
LD_full_df['Incidence'] = LD_full_df.apply(funcInc, axis=1)
```

The number of observations in the full dataset for each class of the Incidence variable is displayed using the code below.

```python
LD_full_df['Incidence'].value_counts()
```

The following function displayed in the code below created the classes for the categorical case count target variable. The function was then applied to the full Lyme disease dataset where the values in the CaseCat column categorically described the Lyme disease case count for the year in each values' respective row.

```python
def funcCaseCat(row):
    if row['CaseCount'] < 10:
        val = 0
    elif row['CaseCount'] < 75:
        val = 1
    elif row['CaseCount'] < 225:
        val = 2
    else:
        val = 3
    return val
```

```python
LD_full_df['CaseCat'] = LD_full_df.apply(funcCaseCat, axis=1)
```

After splitting the data into training features and targets and test features and targets, the training data was resampled. I undersampled the majority Incidence class in the training set so that there would be an equal number of high and low incidence observations during training. When there is a siginificant imbalance of classes during training, some classifiers, such as random forests, may tend to label unseen observations as the majority class simply because when it did that during training, it yielded high results which could lead to not as accurate or strong of results on unseen data. With a K-Nearest Neighbors classifier, if there is significantly more of one class in the training set, there might be more nearest neighbors to a test sample point that are of that one class because there is just more of that one class to begin with and in majority voting, the test sample would get labeled as the class that has the most representatives within the nearest neighbors of the test point.

- Proportion 2 represents the proportion of the number of samples (value in "Number of Samples" column) out of the total number of samples in its respective incidence class (sum of the values in the "Number of Samples" column of the same "Future Incidence" class). 
- Proportion 3 represents the proportion of the number of samples (value in "Number of Samples" column) out of the total number of samples in its respective case count class (sum of the values in the "Number of Samples" column of the same "Future CaseCat" class).

#<h1 align="center">Chapter 4 <br> Machine Learning and Classification</h1>

## 4.1 Preperation

Before classifying the data, it was important to know what type of classification I was doing.

Binary classification is when each sample is labeled as only one class out of only two classes for a target variable (e.g. Positive or negative usually represented by 0 or 1 respectively).

Multiclass classification is when each sample is labeled as only one class out of more than two classes for a target variable. For example, if classifying images that display only one type of clothing item, either a shirt, pants, dress, skirt, or shoes, into one of the type of clothing classes (shirt, pants, dress, skirt, shoes), then that would be a multiclass classification.

Multilabel classification is when each sample is classified as a set of labels which are not mutually exclusive where the set can include a number of labels from 0 to the number of classes. For example, when classifying images taken in a clothing store, the images might display a shirt, pants, dress, skirt, or shoes, or any set or those, or none of them. The classification of labeling what is displayed in the images as exampled is multilabel classification.

Multioutput-multiclass classification, or multi-task classification, is when there is more than one output, or target variable, and a target variable has more than two classes. For example, it could be classifying clothing images by type of clothing item and age group where each image displays one type of clothing item out of the five classes shirt, pants, dress, skirt, and shoes and each clothing item is suited for one of the age group classes infant, child, or adult.

My classification consisted of two target variables: Lyme disease incidence and case count category. Since one of my target variables, case count category, had more than two classes, Scikit-learn treated my classification as a Multioutput-multiclass classification. Using Scikit-learn's classifiers, a valid representation of multioutput y in which the class labels were displayed was a dense matrix of shape (number of samples, number of target variables). Thus, my target data was transformed to match this format as presented in chapter section .



Multioutput-multiclass, also known as multitask, classification was done. ###  4.2.5 Recall and Precision

Recall and precision consider true and false positives and negatives. In the equations shown below, a true positive is considered a sample that the classifier labeled as the specified positive class and its true label was indeed the specified positive class; a false positive is considered a sample that the classifier labeled as the specified positive class, but its true label was not the specified positive classl; a true negative is considered a sample that the classifier labeled as a class that was not the specified positive class and its true label was indeed not the specified positive class; a false negative is considered a sample that the classifier labeled as a class that was not the specified positive class, but its true label was the specified positive class.

Here are the equations for recall and precision:
<center> $TP = \text{True positives}$ </center>
<center> $FP = \text{False positives}$ </center>
<center> $TN = \text{True negatives}$ </center>
<center> $FN = \text{False negatives}$ </center>
<center> $\text{Precision} = TP/(TP+FP)$ </center>
<center> $\text{Recall} = TP/(TP+FN)$ </center>

Recall is the accuracy of the data for one class. For example, if tested the subset consisting of only the observations which had high incidence, recall for high incidence is same as the accuracy of that subset.

In example, a court judge might want high precision since innocent till proven guilty, while a lab test or car organization might want high recall to be better safe than sorry (catch all the positive cases for a disease or catch all the vehicles that fail to meet minimum safety standards).

Given that the confusion matrix for Incidence is $\hspace{.2cm}$
[[4619  427] $\hspace{.2cm}$ it follows that \
$\hspace{9.2cm}$ [ 479 4567]],

recall for the high incidence class is:
$$\frac{4619}{4619+427}=0.9153785176377328$$

and recall for the low incidence class is:
$$\frac{4567}{4567+479}=0.9050733254062624.$$

The macro average of recall (or precision) is the average of the recall (or precision) scores. The micro average of recall (or precision) is the weighted average of the recall (or precision) scores. The macro recall and precision for low incidence and the micro recall for high incidence are shown below.

```python
from sklearn.metrics import precision_score, recall_score

recall_score(y_train_us_enc_shuf[:,0], y_train_pred_rf0, average='macro')
```

The function below combined the two outputs to form one output where each class in the output was a represention for the incidence and case count category outputs. More explicitly, labels that were low incidence and case count class 0 were given a label of 0; labels that were low incidence and case count class 1 were given a label of 1; labels that were low incidence and case count class 2 were given a label of 2; labels that were low incidence and case count class 3 were given a label of 3; labels that were high incidence and case count class 0 were given a label of 4; labels that were high incidence and case count class 1 were given a label of 5; labels that were high incidence and case count class 2 were given a label of 6; labels that were high incidence and case count class 3 were given a label of 7. 

```python
def funcCombLabs(row):
    if row[0] == 0:
        if row[1] == 0: 
          val = 0
        elif row[1] == 1: 
          val = 1
        elif row[1] == 2: 
          val = 2
        elif row[1] == 3: 
          val = 3
    else:
        if row[1] == 0: 
          val = 4
        elif row[1] == 1: 
          val = 5
        elif row[1] == 2: 
          val = 6
        elif row[1] == 3: 
          val = 7
    return val

y_train_pred_rf_df['CombLabels'] = y_train_pred_rf_df.apply(funcCombLabs, axis=1)
np.ravel(y_train_pred_rf_df['CombLabels'])
```

##  4.3 K-Nearest Neighbors

### 4.3.1 Description and Standardization

In the K-Nearest Neighbors (KNN) algorithm each training sample's feature data is made into a vector in a multidimensional space. During training, the algorithm stores the training samples feature vectors with their corresponding class labels. During classification, each test sample feature data are made into vectors and put into the multidimenional feature space where each test sample is classified based on the labels of the k closest training sample vectors.   

When the weight function in Scikit-learn's KNN classifier is `uniform`,  all sample points in each neighborhood are weighted equally so the test sample is classified by assigning the label that is most frequent among the k training samples nearest to that test point (nearest neighbors). When the weight function in Scikit-learn's KNN classifier is `distance`,  the sample points or vectors are weighted by the inverse of their respective distance to the test point or vector. [33] With this, closer neighbors, or neighbors with a smaller distance, to the test point have a greater influence than the neighbors that are further away.

The distance between two points in the multidimensional feature space is the distance between two vectors and those two vectors each contain the feature data values of their own sample. Hence, if the range of values of one feature is much different than another feature, the distance calculations will be off. Thus, each feature data was standardized before being put into the KNN classifier.


To standardize each feature, I used `StandardScaler` which transformed the data such that the distribution for each feature had a mean value 0 and standard deviation of 1.

## 4.5 Ensemble

### 4.5.1 Voting Classifier

The ensemble method used was a soft voting classification. Soft voting classification involves the prediction of the class labels based on the argument of the maximum of the sums of the predicted probabilities. In this case, the probabilites were from the Random Forest classifier and K-Nearest Neighbors classifier from the grid-searches. For random forests, "the predicted class probabilities of an input sample are computed as the mean predicted class probabilities of the trees in the forest. The class probability of a single tree is the fraction of samples of the same class in a leaf" [32]. For KNN, the probability estimates of each class are based on the proportion of labels among nearest neighbors and the nearest neighbors distance to the test point when `weights='distance'`.


#### 4.5.1.1 Fitting Voting Classifier

The following code sets up and fits the Soft Voting classifier for both target variables. The input of the Soft Voting classifier is the unscaled data. The Random Forest classifier found in the grid-search takes the unscaled data as input while the KNN classifier is put into a pipeline where the data is first scaled then given to the KNN estimator found in the grid-search as input.

```python
from sklearn.ensemble import VotingClassifier
from sklearn.pipeline import make_pipeline

scale_knn0 = make_pipeline(StandardScaler(), grid_search_knn.best_estimator_.estimators_[0])
rf_clf0 = grid_search_rf.best_estimator_.estimators_[0]

voting_clf0 = VotingClassifier(
    estimators=[('rf', rf_clf0),
                ('knn', scale_knn0)],
    voting='soft')

v0 = voting_clf0.fit(X_train_us_shuf, y_train_us_enc_shuf[:,0])
```

```python
from sklearn.ensemble import VotingClassifier

scale_knn1 = make_pipeline(StandardScaler(), grid_search_knn.best_estimator_.estimators_[1])
rf_clf1 = grid_search_rf.best_estimator_.estimators_[1]

voting_clf1 = VotingClassifier(
    estimators=[('rf', rf_clf1), 
                ('knn', scale_knn1)],
    voting='soft')

v1 = voting_clf1.fit(X_train_us_shuf, y_train_us_enc_shuf[:,1])
```

## 6.2 Further

The incidence target variable was undersampled and it produced good results for both classes "low" and "high". This project could be furthered by oversampling or undersampling the categorical case count target variable as well since it did not produce evenly good results among its classes, specifically the higher classes 2 and 3 did not perform as well.

An addition of feature(s) representing medical or insurance coverage and/or county poverty rates could be explored as the number of reported cases might be less for areas where poeple do not have good health insurance where they could go to the doctor to get tested for Lyme disease. Another feature could be the amount of travel or location of travel the residing poeple of counties and county equivalents take. It is not uncommon to have coinfections of other tick-borne diseases with Lyme disease so more tick-borne diseases could be explored with Lyme disease such as Babesiosis.
